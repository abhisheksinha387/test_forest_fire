```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning

```python
## missing values
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[0:122,'Region']=0
dataset.loc[122:,'Region']=1
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int) 
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## romove the null values
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df.drop([122],inplace=True)
```

```python
df.columns
```

```python
## fix the spaces in the columns names
df.columns = df.columns.str.strip()
df.columns
```

## change the required column as integer data types

```python
df.columns
```

```python
df.info()
```

```python
df.head()
```

```python
df['month'].value_counts()
```

```python
df[['month','day','year','Temperature','RH','Ws']]  = df[['month','day','year','Temperature','RH','Ws']].astype(int)  
```

## changing the other columns to float data types

```python
objects = [features for features in df.columns if df[features].dtypes=='object']
```

```python
for i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
##lets save the cleaned dataset
df.to_csv("Algerian_forest_fires_dataset.csv",index=False)
```
